```python
#how to create your own dataframe

import pandas as pd
import numpy as np

#np.random.seed(24) 
df = pd.DataFrame({'F': np.linspace(1, 10, 10)})  #create a kinda serial number feature

#concatenating another datafram with random values to the previous one
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 5), columns=list('EDCBA'))], axis=1)

#setting the value in row1 cond colum2 to null
df.iloc[0, 2] = np.nan
df
```

```python
"""Function to colour values
*Negative values are RED
*Positive values are BLACK
*Values equal to zero or null are GREEN
"""


def colorNegativeValueToRed(value):
  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'black'
  else:
    color = 'green'

  return 'color: %s' % color
```

```python
#calling the function to the dataframe

s = df.style.applymap(colorNegativeValueToRed, subset=['A','B','C','D','E'])
s
```

```python
#calling the function on a particular feature in the dataframe

d = df.style.applymap(colorNegativeValueToRed, subset=['A'])
d
```

```python
#using background colour instead
def colorNegativeValueToRed(value):
  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'white'
  else:
    color = 'green'

  return 'background-color: %s' % color 
```

```python
o = df.style.applymap(colorNegativeValueToRed, subset=['A','B','C','D','E'])
o
```
